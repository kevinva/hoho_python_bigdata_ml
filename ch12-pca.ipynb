```python
import os
from PIL import Image
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
```

```python
names = os.listdir('./data/olivettifaces')
```

```python
img0 = Image.open('./data/olivettifaces/' + names[0])
img0.show()
```

```python
img0 = img0.convert('L')  # 灰度图
img0 = img0.resize((32, 32))
arr = np.array(img0)
```

```python
pd.DataFrame(arr)
```

```python
X = []
y = []
for i in names:
    img = Image.open('./data/olivettifaces/' + i)
    img = img.convert('L')
    img = img.resize((32, 32))
    arr = np.array(img)
    X.append(arr.reshape(1, -1).flatten().tolist())
    
    y.append(int(i.split('_')[0]))
```

```python
X = pd.DataFrame(X)
X.shape
```

```python
len(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
pca = PCA(n_components=100)
pca.fit(X_train)
```

```python
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
print(X_train_pca.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X_train_pca, y_train)
```

```python
y_pred = knn.predict(X_test_pca)
y_pred_proba = knn.predict(X_test_pca)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_pred, y_test)
```
