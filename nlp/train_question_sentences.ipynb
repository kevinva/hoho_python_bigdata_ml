```python
import json
import random
import pandas as pd
```

```python
webqa_filepath = '/kaggle/input/qadataset/me_train.json'
with open(webqa_filepath, encoding='utf-8') as file:
    webqa_str = file.read()
    webqa_info = json.loads(webqa_str)
```

```python
webqa_q_list = []
webqa_a_list = []

for key, val in webqa_info.items():
    question = val.get('question', 'no_question')
    if question == 'no_question':
        continue
        
    webqa_q_list.append(question)
    
    evidences_info = val.get('evidences')
    if evidences_info is not None:
        keys = list(evidences_info.keys())
        eidx = random.randint(0, len(keys) - 1)
        answer_info = evidences_info.get(keys[eidx])
        if answer_info is not None:
            answer = answer_info.get('evidence')
            if answer is not None:
                answer = answer.replace('答：', '')
                webqa_a_list.append(answer)
```
